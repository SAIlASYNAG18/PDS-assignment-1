```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('clean_data.csv')
#violin plot for lunch distribution by gender
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.violinplot(x='gender', y='lunch', data=df)
plt.title('Violin Plot of Lunch Types by Gender')
plt.show()
```

```python
sns.set(style="whitegrid")
#bargraph for Parental Level of Education distribution
plt.figure(figsize=(14, 8))
sns.barplot(x='parental level of education', y='math score', data=df, ci=None, label='Math')
sns.barplot(x='parental level of education', y='reading score', data=df, ci=None, label='Reading')
sns.barplot(x='parental level of education', y='writing score', data=df, ci=None, label='Writing')
plt.title('Average Scores by Parental Level of Education')
plt.legend(title='Subject')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.set(style="whitegrid")
#pointplot for Score Distribution by Gender
plt.figure(figsize=(12, 8))
sns.pointplot(x='gender', y='writing score', data=df, ci=None, label='Writing')
sns.pointplot(x='gender', y='math score', data=df, ci=None, label='Math')
sns.pointplot(x='gender', y='reading score', data=df, ci=None, label='Reading')
plt.title('Point Plot of Score Distribution by Gender')
plt.legend(title='Subject')
plt.show()
```

```python
sns.set(style="whitegrid")
#count plot for Distribution of Race/Ethnicity by Gender.
plt.figure(figsize=(10, 7))
sns.countplot(x='race/ethnicity', hue='gender', data=df, palette='Set2')
plt.title('Distribution of Race/Ethnicity by Gender')
plt.show()
```

```python
sns.set(style="whitegrid")
# box plot for test preparation by gender
plt.figure(figsize=(12, 8))
sns.boxplot(x='gender', y='math score', hue='test preparation course', data=df)
plt.title('Box Plot of Math Scores by Gender and Test Preparation Course')
plt.show()
```

 A bar graph makes it simple and clear to understand for me to analyize which level of education people getting how many scores in each subject. So that we can easily understand score distribution by parental level of education.
